```python
# Function's plot with extremes marked
x1 = np.sqrt (2/3)
x2 = -np.sqrt(2/3)
ax = plt.figure().add_subplot(1,1,1)
ax.axvline (x=0, color = 'k')
ax.axhline (y=0, color = 'k')
ax.axvline(x=x1, color = 'r')
ax.axvline(x=x2, color = 'r')
plt.plot(X,f(X))
plt.scatter(x1, f(x1), color = 'r')
plt.scatter (x2, f(x2), color = 'r')
plt.show()
```

With the previous plot we can see how the analitical results obtained are congruent. Note that the points marked in red (maxima and minima) are the ones that satisfies $f'(x) = 0$.

3. In order to check whether the points we found are a minimum or a maximum we may compute the 2nd order Taylor's expansion around each point, $x^*$,that is:
$$
f(x^*+ h) \approx f(x^*) + h f'(x^*) + \frac{1}{2}h^2 f''(x^*)
$$
So in order for $x^*$ to be a minimum we want the second derivative of the function at that point to be positive, $f''(x^*) > 0$. Similarly, if $f''(x^*)<0$, we obtain a maximum.


```python
#Definition of the second derivative and evaluation of the points obtained before.
def f2 (x):
  return 6*x

print (f2(x1))
print (f2(x2))
```

Since $f''(x_1)>0$, then the point $ x_1 =\sqrt{\frac{2}{3}}$ is a minimum. Similarly, we have $f''(x_2)<0$, thus  $ x_2 = -\sqrt{\frac{2}{3}}$ is a maximum.

3.
The elements of the Hessian matrix $\nabla ^2 f(x^*)$ can be defined as: $\nabla ^2 f(x^*)_{ij} = \frac{\partial ^2 f(x^*)}{\partial x_i \partial x_j}$, where $1 \leq i \leq j \leq n$.



Note that $f(x_1,x_2)$ can be expressed as the sum of two one-variable functions, $f(x_1, x_2) = f(x_1) + f(x_2)$, which means that the Hessian is a diagonal matrix, and due to its symmetry in $x_1$ and $x_2$, the elements in the diagonal are equal. In this case we have that $ \frac{ \partial ^2 f(x^*)}{ \partial ^2 x_1} = \frac{\partial ^2 f(x^*)}{ \partial ^2 x_2} = 2  $.

So the Hessian matrix is:

$\nabla ^2 f(x^*) = \begin{pmatrix}
2 & 0\\
0 & 2
\end{pmatrix}
$

We want to know now if the previous matrix is strictly positive or strictly negative. It's known that when a matrix is diagonal we can find its eigenvalues in the diagonal. Therefore, in this case, the eigenvalue is 2 with multiplicity equals 2. As all the eigenvalues are positive, we can conclude that $f$ has a minimum in $\textbf{x}^*=(0,0)$.

**2.2 A two dimensional function with multiple minima**

First of all, we define the function that is going to be studied:

$ f(x) = x_1^2(4-2.1x_1^2 + \frac{1}{3}x_1^4)+x_1x_2+x_2^2(-4+4x_2^2)$

```python
### Definition of the function in 2.2
def f22 (x1,x2):
  return x1**2*(4-2.1*x1**2 + (1/3)*x1**4)+ x1*x2 + x2**2*(-4 + 4*x2**2)
```

```python
### Definition of the points that are going to be evaluated
X1_22 = np.arange(-2, 2, 0.1)
X2_22 = np.arange(-1, 1, 0.1)
X1_22, X2_22 = np.meshgrid(X1_22, X2_22)
Z_22 = f22 (X1_22, X2_22)
```

We will plot this function within the range $x_1 \in [-2,2]$ and $x_2 \in [-1,1]$ and using a step of of 0.1. The plot is the following:

```python
### Plotting the function within the intervals defined
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X1_22, X2_22, Z_22, rstride=1, cstride=1, cmap=cm.coolwarm,
                linewidth=0, antialiased=False)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.contourf(X1_22,X2_22,Z_22)
plt.show()
```

2.

Analytically the gradient $\nabla f(x)$ can be expressed as follows:

$\nabla f = ( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2} ) = ( 8x_1-8.4x_1^3+2x_1^5 + x_2, x_1 - 8x_2+16x_2^3 )$



```python
### Definition of the gradient according to the formula
def grad_f22 (x1,x2):
  return 8*x1-8.4*x1**3+2*x1**5 + x2, x1 - 8*x2 + 16*x2**3
```

3.

We will now study numerically an approximation of the points $x^*$ at which $∇f(x^*)=0$.

```python
### Definition of the norm of a vector
def norm (a, b):
  return a*a + b*b
```

```python
X1_grad = np.arange(-2, 2, 0.005)
X2_grad = np.arange(-1, 1, 0.005)
```

```python
### Here we calculate the gradient for each point
grads_matrix = []
for aux1 in X1_grad:
  auxcolumn = []
  for aux2 in X2_grad:
    grad1, grad2 = grad_f22(aux1, aux2)
    auxcolumn.append(norm(grad1, grad2))
  grads_matrix.append(auxcolumn)

# np.shape(grads_matrix) = (800,400)
```

```python
### Compare the gradients obtained with their neighbors,
### and decide if it represents an extreme point
neighbors = [-1, 0, 1]
extreme_gradient = []
for i in range(len(X1_grad)):
  if i== 0 or i==len(X1_grad)-1:
    continue
  for j in range(len(X2_grad)):
    if j== 0 or j==len(X2_grad)-1:
      continue
    flag = 0
    for cont1 in neighbors:
      for cont2 in neighbors:
        if cont1 == 0 and cont2 == 0:
          continue
        #print (i, j, cont1, cont2)
        if grads_matrix[i+cont1][j+cont2] <= grads_matrix[i][j]:
          flag = 1
          break
    if flag == 0:
      auxcolumn = [X1_grad[i], X2_grad[j]]
      extreme_gradient.append(auxcolumn)
```

We obtained 15 points. These points and the norm, $||\nabla f(x,y)||^2$ of them, are the following:

```python
for i in range (len(extreme_gradient)):
  a , b =grad_f22 (extreme_gradient[i][0],extreme_gradient[i][1])
  print("The %d point is (%.3lf, %.3lf) and its norm is %.3le" %(i+1 , extreme_gradient[i][0],extreme_gradient[i][1], norm(a,b)))
```

4.
The elements of the Hessian matrix $\nabla ^2 f(x^*)$ can be defined as: $\nabla ^2 f(x^*)_{ij} = \frac{\partial ^2 f(x^*)}{\partial x_i \partial x_j}, $ where $1 \leq i \leq j \leq n$

Taking that into account we obtain the following: $\nabla ^2 f(x^*) = \begin{pmatrix}
8-25.2x_1^2+10x_1^4 & 1\\
1 & -8+48x_2^2
\end{pmatrix}
$

```python
### Definition of the second derivatives and the Hessian
def f22_second_derivative_x1 (x1, x2):
  return 8-25.2*x1**2+10*x1**4
def f22_second_derivative_x2 (x1, x2):
  return -8 + 48*x2**2
def Hessian (x1, x2):
  H = np.zeros((2,2))
  H [1][0] = H [0][1] = 1
  H [0][0] = f22_second_derivative_x1 (x1, x2)
  H [1][1] = f22_second_derivative_x2 (x1, x2)
  return H
```

```python
### Library that allows us to obtain eigenvalues and eigenvectors
from numpy import linalg as LA
```

We now want to study if the points that vanishes the gradient are minimums, maximums or saddle points. In order to do that, we study whether the Hessian matrix is positive definite, negative definite or neither of the latter, respectively.

```python
### For each coordinates that have a lower gradient than its neighbors, we obtain the eigenvalues based on the Hessian matrix
eigval_list = []
for i in range (len(extreme_gradient)):
  h = Hessian (extreme_gradient[i][0], extreme_gradient[i][1])
  eigenvalues, eigenvectors = LA.eig(h)
  eigval_list.append(eigenvalues)
  print ('The eigenvalues of the Hessian matrix evaluated in the coordinates (%.3lf, %.3lf) are %.3lf and %.3lf '  %(extreme_gradient[i][0],extreme_gradient[i][1], eigenvalues[0], eigenvalues[1]))
```

According to the sign of the eigenvalues, we conclude that:

```python
### Based on the sign of the eigenvalues, we decide if the coordinates correspond to a maximum. a minimum or a saddle point
for i in range (len(extreme_gradient)):
  if eigval_list[i][0] > 0 and eigval_list[i][1] > 0:
    print ('The coordinates (%.3lf, %.3lf) correspond to a minimum' % (extreme_gradient[i][0], extreme_gradient[i][1]))
  elif eigval_list[i][0] < 0 and eigval_list[i][1] < 0:
    print ('The coordinates (%.3lf, %.3lf) correspond to a maximum' % (extreme_gradient[i][0], extreme_gradient[i][1]))
  else:
    print ('The coordinates (%.3lf, %.3lf) correspond to a saddle point' % (extreme_gradient[i][0], extreme_gradient[i][1]))
```
